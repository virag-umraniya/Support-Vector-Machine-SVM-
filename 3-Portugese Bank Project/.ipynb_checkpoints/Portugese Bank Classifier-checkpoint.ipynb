## 3. Data Preprocessing

### i. Checking for NaN/Outliers

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes('number')
num_cols
```

```python
for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=col, data=df)
    plt.show()
```

```python
drop_out_cols = df[['age', 'duration', 'campaign', 'cons.conf.idx']]
drop_out_cols
```

```python
for col in drop_out_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lb = q1 - 1.5 * iqr
    ub = q3 + 1.5 * iqr
    df = df[(df[col] >= lb) & (df[col] <= ub)]
df
```

```python
# Remove rows with any 'unknown' values
df = df[~df.isin(['unknown']).any(axis=1)]
```

```python
df.drop('month', axis=1, inplace=True)
```

### ii. Scaling

```python
df.campaign.value_counts()
```

```python
df.previous.value_counts()
```

```python
# Convert pdays column to binary: 0 = not contacted, 1 = contacted
df['pdays'] = df['pdays'].apply(lambda x: 0 if x == 999 else 1)
df
```

```python
num_cols = df.select_dtypes('number')
num_cols
```

```python
scale_cols = num_cols.drop(['campaign', 'pdays'], axis=1)
```

```python
scale_cols
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled = scaler.fit_transform(scale_cols)
df_scaled
```

```python
df_scaled = pd.DataFrame(df_scaled, columns=scale_cols.columns)
df_scaled
```

```python
df_scaled = pd.concat([df_scaled, df[['campaign', 'pdays']].reset_index(drop=True)], axis=1)
df_scaled
```

### iii. Encoding

```python
cat_cols = df.select_dtypes('object')
cat_cols
```

```python
for col in cat_cols:
    print(df[col].value_counts())
    print('----------------------------------------------------')
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
import pandas as pd

# Step 1: Make a copy of the original DataFrame
df_encoded = cat_cols.copy()

# Step 2: One-hot encode nominal categorical columns
onehot_cols = ['job', 'marital', 'contact', 'day_of_week', 'poutcome']
df_encoded = pd.get_dummies(df_encoded, columns=onehot_cols, drop_first=True)

# Step 3: Ordinal encode 'education' with defined order
education_order = ['illiterate', 'basic.4y', 'basic.6y', 'basic.9y',
                   'high.school', 'professional.course', 'university.degree']
ordinal = OrdinalEncoder(categories=[education_order])
df_encoded['education'] = ordinal.fit_transform(df_encoded[['education']])

# Step 4: Label encode binary columns: 'default', 'housing', 'loan', 'y'
label_cols = ['default', 'housing', 'loan', 'y']
label_encoder = LabelEncoder()
for col in label_cols:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])

# Step 5: Convert True/False (boolean) values to 0/1 if present
df_encoded = df_encoded.applymap(lambda x: int(x) if isinstance(x, bool) else x)
```

```python
df_encoded
```

```python
df_encoded
```

```python
# Reset index (if needed) to align rows
df_scaled = df_scaled.reset_index(drop=True)
df_encoded = df_encoded.reset_index(drop=True)

# Concatenate side-by-side (columns-wise)
df_final = pd.concat([df_scaled, df_encoded], axis=1)
df_final
```

```python

# Correlation matrix
corr_matrix = df_final.corr().abs()

# Target column correlation
target_corr = corr_matrix['y']

# Drop target from feature-feature matrix to avoid confusion
corr_matrix_features = corr_matrix.drop('y').drop('y', axis=1)

# Upper triangle to avoid duplicates
upper = corr_matrix_features.where(
    np.triu(np.ones(corr_matrix_features.shape), k=1).astype(bool)
)

# Threshold for high correlation
threshold = 0.9

# Print highly correlated feature pairs with their correlation to the target
print("Highly Correlated Feature Pairs (Correlation > 0.9):\n")
for col1 in upper.columns:
    for col2 in upper.columns:
        if upper.loc[col1, col2] > threshold:
            print(f"{col1} and {col2} => Corr: {upper.loc[col1, col2]:.2f} | "
                  f"{col1} vs y: {target_corr[col1]:.2f}, {col2} vs y: {target_corr[col2]:.2f}")
            print('-------------------------------------------------------------------------------------------------')
```

```python
df_final.drop('euribor3m', axis=1, inplace=True)
df_final
```

## 5. Feature Extraction

```python
X = df_final.drop('y', axis=1)
X
```

```python
y = df_final[['y']]
y
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc = pca.fit_transform(X)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number Component')
plt.ylabel('Variance in %')
plt.title('Explained Variance')
plt.show()
```

```python
np.where(np.cumsum(pca.explained_variance_ratio_) <= 0.90)
```

```python
pca = PCA(n_components=10)
new_data = pca.fit_transform(X)
new_data
```

```python
# column names
columns = [f'Principle Component {i}' for i in range(1,11)]
# creating df
principle_df = pd.DataFrame(data=new_data, columns = columns)
principle_df
```

```python
principle_df = pd.concat([principle_df, y.reset_index(drop=True)], axis=1)
principle_df
```

```python
## 
```

```python
X = principle_df.drop('y', axis=1)
X
```

```python
y = principle_df[['y']]
y
```

## 6. Split the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## 7. Fit the data to the model

```python
principle_df.y.value_counts()
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Predict on training and test set
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Classification report on training data
print("=== Classification Report on Training Data ===")
print(classification_report(y_train, y_train_pred))

# Classification report on test data
print("=== Classification Report on Test Data ===")
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.model_selection import cross_val_score

# Cross-validation (5-fold)
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

# Results print
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
print("Standard deviation:", scores.std())
```

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Step 1: Train SVC with probability=True
model = SVC(probability=True)
model.fit(X_train, y_train)

# Step 2: Get predicted probabilities (positive class only)
y_proba = model.predict_proba(X_test)[:, 1]

# Step 3: Calculate ROC AUC
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc = roc_auc_score(y_test, y_proba)

# Step 4: Plot ROC curve
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"SVC (AUC = {auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - SVC")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

## Data Balancing

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

# Optional: Check new class distribution
print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_bal.value_counts())
```

```python
from sklearn.svm import SVC
model = SVC(probability=True)
model.fit(X_train_bal, y_train_bal)
```

```python
from sklearn.metrics import classification_report

# Predict on training and test set
y_train_pred = model.predict(X_train_bal)
y_test_pred = model.predict(X_test)

# Classification report on training data
print("=== Classification Report on Training Data ===")
print(classification_report(y_train_bal, y_train_pred))

# Classification report on test data
print("=== Classification Report on Test Data ===")
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.model_selection import cross_val_score

# Cross-validation (5-fold)
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')

# Results print
print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
print("Standard deviation:", scores.std())
```
